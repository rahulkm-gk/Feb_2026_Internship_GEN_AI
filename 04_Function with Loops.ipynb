# Problem 1: Mobile Recharge Validation System

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        return "Recharge amount must be at least â‚¹50."
    
    if amount not in valid_plans:
        return "Invalid plan selected. Choose from 199, 299, 399, 599."
    
    return "Recharge successful!"

while True:
    try:
        amount = int(input("Enter recharge amount: "))
        
        result = validate_recharge(amount)
        print(result)
        
        if result == "Recharge successful!":
            break  
    except ValueError:
        print("Please enter a valid numeric amount.")
```

# Problem 3: Student Result Processing System

```python
def check_result(marks):
    total = 0
    for mark in marks:
        total += mark
    avg = total / len(marks)
    if avg >= 50:
        return "Pass"
    else:
        return "Fail"

marks = [90, 80, 60, 40]

result = check_result(marks)

print(result)
```

# Problem 4: Cab Fare Estimator with Retry Option


```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if peak == "yes":
        fare = fare + (fare * 0.25)
    
    return fare
distance = float(input("Enter distance: "))
peak = input("Is it peak hour? (yes/no): ").lower()
calculate_fare_t = calculate_fare(distance, peak)  
print("Total", calculate_fare_t)
```

# Problem 5: Employee Attendance Eligibility Checker

```python
def check_eligibility(attendance):
    present_count = 0
    for day in attendance:
        if day == "P":
            present_count += 1
    percentage = (present_count / len(attendance)) * 100
    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance_list = ["P", "P", "A", "P", "P", "A", "P", "P"]
result = check_eligibility(attendance_list)
print("Result:", result)
```

# Problem 6: Password Strength Checker

```python
# def check_password(passowrd):
#     for char in passowrd:
#         if len(passowrd) >= 8 :
#          return "Strong Password"
#     else:
#       return "Weak Password"
# passowrd = input("Enter any password: ")
# result = check_password(check_password)
# print(result)
def check_password(password):
    has_digit = False
    has_special = False
    
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True
    
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


password = input("Enter any password: ")
result = check_password(password)

print(result)
```
