1. User Login Check
Problem Statement:
Given a username and password, check whether login is successful.
username = "admin"
password = "1234"


```python
username = "admin"
password = "1234"

in_user = input("Please enter your username: ")
in_pass = input("Please enter your password: ")

if in_user == username and in_pass == password :
    print ("login Successful")
else :
    print("Not valid password")
```

2. Pass / Fail Analyzer
Problem Statement:

marks = [45, 78, 90, 33, 60]


```python
marks = [45,78,90,33,60]
for i in marks:
    if (i>=50):
        print("pass")
    else :
        print("fail")
```

5. Error Message Detector
Problem Statement:
Detect error messages from system logs.
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

Requirements:
Count the number of "ERROR" entries


```python
logs = ["INFO","ERROR","WARNING","ERROR"]
total_count = 0;
for i in logs:
    if i == "ERROR":
       
        print (i)
        total_count +=1
    else:
        print("Not ERROR")

print (total_count)
  
```
