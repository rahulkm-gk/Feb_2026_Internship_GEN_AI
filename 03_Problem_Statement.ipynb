#  Statement 2: Search Query Keyword Analysis


```python
import string
sentance = input("Enter any Sentence")
input_lower = sentance.lower()
#LOWER
print(input_lower)
# spt = input_lower.split()
#remove punctuation 
punctuation = input_lower.translate(str.maketrans(" "," ", string.punctuation))
words = punctuation.split()
frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1
result = {words: count for words, count in frequency.items() if count > 1}
print(result)
# remove_punctuation = set(input_lower)
# print(remove_punctuation)

```

# Problem Statement 4: Email Domain Usage Analysis


```python
Input_emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]
domain_count = {}
for email in Input_emails:
    domain = email.split("@")[1]  
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_user = len(Input_emails)

for domain, count in domain_count.items():
    percentage = (count / total_user) * 100
    print(f"{domain}: {int(percentage)}%")
```

# Problem Statement 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average = sum(sales) / len(sales)
above_average= average * 1.3
```

```python
for day , value in enumerate(sales, start=1):
    if value >= above_average:
        print(f"{day}:{value}")
```
