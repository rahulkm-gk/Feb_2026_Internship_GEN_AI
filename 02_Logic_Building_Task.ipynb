# 2: Highest Salary from Employee Data


```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
```

```python
highest_emp = max(employees, key=employees.get)
highest_salary = employees[highest_emp]
print(highest_emp,":",highest_salary)
```

#  4: Count Products Above a Price Threshold


```python
prices = [450, 1200, 899, 1500, 300]
count = 0
for price in prices:
    if price > 1000:
        print(price)
        count +=1
print("Products above 1000",count)
```

# 5: Calculate Attendance Percentage


```python
attendance = ["P", "P", "A", "P", "P"]
total = len(attendance)
print(total)
total_p = attendance.count("P")
print((total_p / total) * 100)
```
